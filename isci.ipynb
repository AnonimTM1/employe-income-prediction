```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("isci_bilgileri.csv")
data.head()
```

```python

x = data['tejribe'].values.reshape(-1,1)
y = data['aylyk'].values.reshape(-1,1)
plt.scatter(x,y)
plt.xlabel("Tejribe")
plt.ylabel("Aylyk")
plt.show()
data.head()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=33)
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
yhead = lr.predict(ytest)
yhead
```

```python
lr.predict([[6]])
```

```python
plt.scatter(x,y)
plt.plot(xtest,lr.predict(xtest),color="red")
```
